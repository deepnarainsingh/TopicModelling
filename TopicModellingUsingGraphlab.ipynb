Using GraphLab Library for Topic Modelling on NY Times articles

```python
'''Using pyLDAvis which is a library for Topic Model Visulisation
Reference taken from the below link
https://github.com/bmabey/pyLDAvis'''
```

```python
import graphlab as gl
import pyLDAvis 
import pyLDAvis.graphlab
import pandas as pd
import ftfy
```

```python
#Automatic Rendering of Visualisation

pyLDAvis.enable_notebook()
```

```python
nyArticlesDF= pd.read_pickle("data/articles.pkl")
```

```python
#Cleaning Data

nyArticlesDF['content'] = nyArticlesDF['content'].apply(lambda x: ftfy.fix_encoding(x)
                                        if isinstance(x, unicode)
                                        else "Warning: not Unicode")
```

```python
#There is one row which is not unicoded
print("{} row isn't unicode".format(sum(nyArticlesDF['content'] == "Warning: not Unicode")))
```

```python
#Dropping the row which is not unicoded

nyArticlesDF = nyArticlesDF[nyArticlesDF['content'] != "Warning: not Unicode"]
```

```python
print("{} row isn't unicode".format(sum(nyArticlesDF['content'] == "Warning: not Unicode")))
```

```python
# Passing the Data Frame from pandas as input to GraphLab SFrame

nyArticlesSFrame= gl.SFrame(nyArticlesDF)
```

```python
nyArticlesSFrame.show()
```

```python
'''Constructing Bag-of-Words Classifier
One of the common techniques to perform document classification (and reviews classification) 
is using Bag-of-Words model, in which the frequency of each word in the document is used 
as a feature for training a classifier. GraphLab's text analytics toolkit makes it easy 
to calculate the frequency of each word in each review. Namely, by using the count_ngrams 
function with n=1, we can calculate the frequency of each word in each review. '''
```

```python
nyArticlesSFrame['features'] = gl.text_analytics.count_ngrams(nyArticlesSFrame ['content'],1)
```

```python
nyArticlesSFrame.show(['content','features'])
```

```python
#Creating Topic Model
#gl.topic_model.create?
topicModel=gl.topic_model.create(nyArticlesSFrame['features'],num_topics=10,num_iterations=50)
```

Visualsation Of Topic Model
---

```python
pyLDAvis.graphlab.prepare(topicModel, nyArticlesSFrame['features'])
```
